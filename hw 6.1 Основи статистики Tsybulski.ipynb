**Завдання 5**. Ми провели дослідження віку студентів-першокурсників. Виявилось, що
- 10 студентів мають вік 16 років
- 25 студентів - 17 років
- 20 студентів - 18 років
- 2 студенти - 20 років

Обчислість середнє значення для цієї вибірки будь-яким способом - на чистому Python або з numpy - та виведіть результат на екран.

```python
mean_age = sum([16 * 10, 17 * 25, 18 * 20, 20 * 2]) / (10 + 25 + 20 + 2)
mean_age
```

```python
ages = [16, 17, 18, 20]
students = [10, 25, 20, 2]
total_students = sum(students)
weighted_sum = sum(age * count for age, count in zip(ages, students))
mean_age = weighted_sum / total_students
print(f"Середнє значення віку: {mean_age}")
```

```python
ages = np.array([16, 17, 18, 20])
students = np.array([10, 25, 20, 2])
mean_age = np.average(ages, weights=students)
print(f"Середнє значення віку: {mean_age}")
```

**Завдання 6**. Крім даних про суму покупок, ми зібрали ще дані про кількість відвідувань кожного з кожного покупця, які знаходяться в змінній `visits`. Обчисліть коефіцієнт кореляції Пірсона для суми покупок і кількості відвідувань кожним з зазначених методів:  

  6.1. без використання жодних бібліотек, користуючись тільки Python.  
  6.2. використовуючи відповідний метод бібліотеки numpy.  
  
Виведіть результат на екран. Проаналізуйте результат.

Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

6.1. без використання жодних бібліотек, користуючись тільки Python.

```python
mean_vis = sum(visits) / len(visits)
mean_vis
```

```python
sum_purch_vis = [(x - mean_sum_purch) * (y - mean_vis) for x, y in zip(sum_purchases, visits)]
cov_purch_vis = sum(sum_purch_vis) / (len(sum_purchases) - 1)
cov_purch_vis
```

```python
std_sum_purchases_sample
```

```python
visits_minus_mean_vis = [(el - mean_vis)**2 for el in visits]
std_visits_sample = np.sqrt(sum(visits_minus_mean_vis)/(len(visits)-1))
std_visits_sample
```

```python
pearson_correlation = cov_purch_vis / (std_sum_purchases_sample * std_visits_sample)
print(f"Коефіцієнт кореляції Пірсона: {pearson_correlation}")
```

6.2. використовуючи відповідний метод бібліотеки numpy.

```python
corr_matrix = np.corrcoef(sum_purchases, visits)
pearson_correlation_np = corr_matrix[0, 1]
print(f"Коефіцієнт кореляції Пірсона: {pearson_correlation_np}")
```

Результат коефіцієнта кореляції Пірсона, що дорівнює -0.3027, вказує на те, що між сумою покупок і кількістю відвідувань існує слабка негативна лінійна залежність.

Негативна кореляція: Значення коефіцієнта кореляції від'ємне, що свідчить про те, що коли кількість відвідувань збільшується, сума покупок схильна дещо зменшуватися, і навпаки. Однак, оскільки коефіцієнт кореляції близький до -0.3, це вказує на досить слабку залежність.

Слабка залежність: Коефіцієнт Пірсона близький до -0.3 свідчить про слабкий зв'язок між двома змінними. Це означає, що існує незначний зворотний зв'язок, але він не є сильним або чітко вираженим. Іншими словами, на підставі цих даних неможливо точно передбачити одне значення за іншим.

```python
import matplotlib.pyplot as plt

plt.scatter(sum_purchases, visits)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

Висновок:
Коефіцієнт кореляції Пірсона -0.3027 підтверджує, що між кількістю відвідувань і сумою покупок є слабкий негативний зв'язок.
Кореляція не є лінійною - якби дані мали більш виражену лінійну залежність, коефіцієнт кореляції був би ближчим до -1 або 1. У цьому випадку результат вказує на слабкий зв'язок.
До того ж присутні викиди, які абсолютно не лінійні й негативно впливають на кореляцію, роблячи її менш лінійною.

**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
threshold = mean_sum_purch + 2 * std_sum_purchases_sample
print(f"Поріг для викидів (середнє + 2 стандартних відхилення): {threshold}")
```

```python
filtered_sum_purchases = [x for x in sum_purchases if x <= threshold]
filtered_visits = [visits[i] for i in range(len(sum_purchases)) if sum_purchases[i] <= threshold]
```

```python
print(filtered_sum_purchases)
print(filtered_visits)
```

```python
corr_matrix = np.corrcoef(filtered_sum_purchases, filtered_visits)
pearson_correlation = corr_matrix[0, 1]
print(f"Коефіцієнт кореляції Пірсона після вилучення викидів: {pearson_correlation}")
```

```python
plt.scatter(filtered_sum_purchases, filtered_visits)
plt.title('Scatter Plot of Purchases vs. Visits (without outliers)')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits')
plt.show()
```

Результат коефіцієнта кореляції Пірсона -0.3685 після вилучення викидів вказує на слабку негативну кореляцію між сумою покупок і кількістю відвідувань, але на цей раз зв'язок став ще менш вираженим порівняно з попереднім результатом -0.3027, коли викиди були присутні.
Після вилучення викидів коефіцієнт став більш від'ємним (вказує на деяке посилення негативної кореляції), але залишається слабким, також збільшення машстаба графіка показує, що дані не розташовані в якійсь лінійній залежності відносно один до одного.
Це може свідчити про те, що викиди впливали на початкову оцінку, але, навіть після їх вилучення, зв'язок залишається слабким.

Хоча викиди були вилучені, це не змінило суттєво саму природу взаємозв'язку між змінними, що може вказувати на те, що інші фактори можуть впливати на ці змінні або вони мають нелінійний характер взаємодії.
